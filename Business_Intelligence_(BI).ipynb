```python
# Importação das bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Criação de um DataFrame de exemplo com anos e vendas
data = {
    'Ano': range(2025, 2035),
    'Vendas': [100, 150, 200, 250, 300, 350, 400, 450, 500, 550]
}
df = pd.DataFrame(data)
print(df)

# Gráfico de Vendas Anuais
plt.figure(figsize=(10, 6))
plt.plot(df['Ano'], df['Vendas'], marker='o')
plt.title('Vendas Anuais')
plt.xlabel('Ano')
plt.ylabel('Vendas')
plt.grid(True)
plt.show()

# Interpretação: Este gráfico mostra o aumento constante nas vendas ao longo dos anos, indicando um crescimento positivo.

# Gráfico de Pizza para Participação Anual nas Vendas
plt.figure(figsize=(8, 8))
plt.pie(df['Vendas'], labels=df['Ano'], autopct='%1.1f%%', startangle=140)
plt.title('Participação Anual nas Vendas')
plt.show()

# Interpretação: Este gráfico de pizza destaca a contribuição de cada ano para o total das vendas, ilustrando a crescente participação ao longo do tempo.

# Boxplot de Vendas
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Vendas'])
plt.title('Boxplot de Vendas')
plt.xlabel('Vendas')
plt.grid(True)
plt.show()

# Interpretação: O boxplot mostra a distribuição das vendas, destacando os valores mínimos, máximos, a mediana e os quartis. Isso facilita a visualização de possíveis outliers.

# Comparação entre Modelos: Regressão Linear
# Treinamento do Modelo

# Variáveis independentes e dependentes
X = df[['Ano']]
y = df['Vendas']

# Treinamento do modelo
model = LinearRegression()
model.fit(X, y)

# Previsões
df['Previsoes'] = model.predict(X)
print(df)

# Gráfico de Regressão Linear
plt.figure(figsize=(10, 6))
plt.plot(df['Ano'], df['Vendas'], marker='o', label='Vendas Reais')
plt.plot(df['Ano'], df['Previsoes'], linestyle='--', label='Previsões')
plt.title('Regressão Linear - Vendas Reais vs. Previsões')
plt.xlabel('Ano')
plt.ylabel('Vendas')
plt.legend()
plt.grid(True)
plt.show()

# Interpretação: O gráfico de regressão linear mostra a linha de tendência das vendas, destacando que o modelo está acompanhando bem o crescimento das vendas ao longo dos anos.

# Projeções de 10, 20 e 30 Anos
anos_futuros = pd.DataFrame({'Ano': range(2035, 2065, 10)})
anos_futuros['Previsoes'] = model.predict(anos_futuros)

print(anos_futuros)

plt.figure(figsize=(10, 6))
plt.plot(df['Ano'], df['Vendas'], marker='o', label='Vendas Reais')
plt.plot(df['Ano'], df['Previsoes'], linestyle='--', label='Previsões')
plt.plot(anos_futuros['Ano'], anos_futuros['Previsoes'], linestyle='--', marker='o', label='Projeções Futuras')
plt.title('Projeções de Vendas para os Próximos 10, 20 e 30 Anos')
plt.xlabel('Ano')
plt.ylabel('Vendas')
plt.legend()
plt.grid(True)
plt.show()

# Interpretação: As projeções indicam um crescimento contínuo das vendas nos próximos 10, 20 e 30 anos, seguindo a tendência observada nos dados históricos.

# Gráfico de Dispersão com Linha de Regressão
plt.figure(figsize=(10, 6))
sns.regplot(x=df['Ano'], y=df['Vendas'], ci=None, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.title('Gráfico de Dispersão com Linha de Regressão')
plt.xlabel('Ano')
plt.ylabel('Vendas')
plt.grid(True)
plt.show()

# Interpretação: O gráfico de dispersão com linha de regressão mostra a relação entre o ano e as vendas, indicando uma forte correlação positiva. A linha de regressão ajuda a visualizar a tendência geral dos dados.
```

**Gráfico de Vendas Anuais**

Este gráfico mostra como as vendas aumentaram de forma constante ao longo dos anos. Cada ponto representa as vendas de um ano específico, e a linha conecta esses pontos para destacar a tendência de crescimento. Podemos ver claramente que, a partir de 2025, as vendas aumentam de maneira progressiva até 2034. Isso sugere que a empresa está crescendo de forma estável e contínua.

**Gráfico de Pizza para Participação Anual nas Vendas**

Este gráfico de pizza divide as vendas totais entre os anos de 2025 a 2034. Cada "fatia" representa a porcentagem das vendas em um ano específico. Podemos ver que as fatias vão ficando maiores à medida que os anos passam, indicando que as vendas estão aumentando a cada ano. Isso mostra que a contribuição de cada ano para o total de vendas está crescendo, o que é um sinal positivo para a empresa.

**Boxplot de Vendas**

O boxplot mostra a distribuição das vendas ao longo dos anos. A linha no meio da caixa representa a mediana das vendas. As extremidades da caixa mostram o primeiro e o terceiro quartis (Q1 e Q3), e as "linhas" (chamadas de bigodes) se estendem aos valores mínimo e máximo. Esse gráfico facilita a visualização dos dados e identificação de possíveis outliers (valores extremos). Podemos ver que não há outliers significativos e que as vendas estão bem distribuídas ao longo dos anos.

**Gráfico de Regressão Linear - Vendas Reais vs. Previsões**

Este gráfico compara as vendas reais com as previsões geradas pelo modelo de regressão linear. Os pontos azuis representam as vendas reais, enquanto a linha tracejada vermelha mostra as previsões do modelo. Podemos ver que a linha de previsão acompanha bem os pontos de vendas reais, indicando que o modelo está capturando a tendência de crescimento das vendas.

**Projeções de Vendas para os Próximos 10, 20 e 30 Anos**

Este gráfico mostra as previsões de vendas para os próximos 10, 20 e 30 anos. As linhas tracejadas vermelhas representam as previsões do modelo para os anos futuros. Podemos ver que as projeções indicam um crescimento contínuo das vendas, seguindo a tendência observada nos dados históricos. Isso sugere que, se a empresa mantiver o desempenho atual, as vendas continuarão a aumentar no futuro.

**Gráfico de Dispersão com Linha de Regressão**

Este gráfico de dispersão mostra a relação entre o ano e as vendas, com uma linha de regressão indicando a tendência geral. Os pontos azuis representam as vendas reais em cada ano, enquanto a linha vermelha mostra a linha de regressão ajustada. A forte inclinação positiva da linha de regressão indica uma correlação positiva entre o ano e as vendas, ou seja, as vendas aumentam à medida que os anos passam.
